```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importació de les dades

# Configurem l'estil per visualitzar correctament l'espectre
plt.style.use('seaborn-v0_8-darkgrid')
# Amb aquest estil podem 

# Carreguem les dades respectant l'arquitectura del mode collect
ruta_dataset = '~/Documents/Hyperspectral_camera/Dataset_Entrenament_Nou.csv'
df = pd.read_csv(ruta_dataset, sep=';', decimal=',')

# Extracció del Target_ML (primera lletra de la Label) i separació de les 512 bandes
df['Target_ML'] = df['Label'].astype(str).str[0]
bandes = [col for col in df.columns if col not in ['Label', 'Target_ML']]

# Verificació de valors nuls (ha de ser 0% segons l'especificació)
print(f"Valors nuls detectats: {df.isnull().sum().sum()}")

Estudi de la Distribució i Visualització de Signatures

```python
def plot_signatures_crues(df, bandes):
    plt.figure(figsize=(14, 6))
    colors = sns.color_palette("tab10", len(df['Target_ML'].unique()))
    
    for idx, target in enumerate(df['Target_ML'].unique()):
        subset = df[df['Target_ML'] == target][bandes]
        
        # Calculem la mitjana per classe
        mitjana = subset.mean(axis=0)
        
        # Convertim el nom de les bandes (ex: '325.3') a float per a l'eix X
        longituds_ona = np.array([float(b) for b in bandes])
        
        plt.plot(longituds_ona, mitjana, label=f'Classe {target}', color=colors[idx])

    plt.axhline(0, color='red', linestyle='--', alpha=0.5, label='Límit de Soroll Tèrmic (< 0)')
    plt.title("Signatures Espectrals Crues (VIS + NIR) sota llum Halògena")
    plt.xlabel("Longitud d'ona (nm)")
    plt.ylabel("Reflectància Relativa")
    plt.legend()
    plt.show()

plot_signatures_crues(df, bandes)
```

Correlacions

```python
# Analitzem només un subconjunt (ex: cada 20 bandes) per no col·lapsar la memòria de Jupyter
subset_bandes = bandes[::20]
correlacio = df[subset_bandes].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlacio, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Matriu de Correlació entre longituds d'ona (Mostreig)")
plt.show()
```
